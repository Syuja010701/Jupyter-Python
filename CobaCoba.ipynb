```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

%matplotlib inline
# Generate universe variables

x_kemasan = np.arange(0, 1200, 1)
x_proses = np.arange(0, 12, 1)
x_harga  = np.arange(0, 310000, 1)

# Generate fuzzy membership functions
kemasan_kc = fuzz.trapmf(x_kemasan, [0, 0, 100, 1000])
kemasan_bs = fuzz.trapmf(x_kemasan, [100, 1000, 1200, 1200])

proses_cp = fuzz.trapmf(x_proses, [0, 0, 1, 10])
proses_lb = fuzz.trapmf(x_proses, [1, 10, 12, 12])

harga_rd  = fuzz.trapmf(x_harga, [0, 0, 30000, 300000])
harga_tg  = fuzz.trapmf(x_harga, [30000, 300000, 320000, 320000])


# Visualize these universes and membership functions
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_kemasan, kemasan_kc, 'b', linewidth=1.5, label='kecil')
ax0.plot(x_kemasan, kemasan_bs, 'g', linewidth=1.5, label='besar')

ax0.set_title('Kemasan')
ax0.legend()

ax1.plot(x_proses, proses_cp, 'b', linewidth=1.5, label='cepat')
ax1.plot(x_proses, proses_lb, 'g', linewidth=1.5, label='lambat')


ax1.set_title('Proses')
ax1.legend()

ax2.plot(x_harga, harga_tg, 'b', linewidth=1.5, label='Tinggi')
ax2.plot(x_harga, harga_rd, 'g', linewidth=1.5, label='Rendah')

ax2.set_title('harga')
ax2.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```
