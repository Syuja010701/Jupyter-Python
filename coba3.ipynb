# IMPLEMTASI FUZZY TSUKAMOTO KOPI GAYO PADA JUPYTER PYTHON

Implementasi metode Fuzzy Logic Tsukamoto pada harga Kopi Gayo Arabika Specialty dengan input yang 
di dapatkan Kemasan dan Proses untuk menentukan harga berdasarkan 4 Rules.

a. Rata – rata (avarage)
     Ukuran kemasan paling kecil 100 gram dan paling besar 1000gram
     Proses pembuatan paling cepat 1 hari dan paling lambat 10 hari
     Untuk harga yang murah 30.000 dan paling mahal 30.000
    
    CONTOH KASUS :
    Berapa Harga Kopi Gayo Jika Ukuran kemasan 400Gram Dengan Proses 6 Hari ?
    
b. Tahap FuzzyFikasi

    1. Ukuran Kemasan


```python
import sys
print(sys.version_info)
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
kemasan = 400
x_kemasan = np.arange(100, 1000, 1)

kemasan_kc = fuzz.trapmf(x_kemasan, [100, 100, 100, 1000])
kemasan_bs = fuzz.trapmf(x_kemasan, [100, 1000, 1000, 1000])

fig, ax = plt.subplots(nrows=1, figsize=(7, 4))

ax.plot(x_kemasan, kemasan_kc, 'b', linewidth=1.5, label='kecil')
ax.plot(x_kemasan, kemasan_bs, 'g', linewidth=1.5, label='besar')

ax.set_title('Kemasan')
ax.legend()

plt.tight_layout()

```

```python
in1 = []
in1.append(fuzz.interp_membership(x_kemasan, kemasan_kc, kemasan))
in1.append(fuzz.interp_membership(x_kemasan, kemasan_bs, kemasan))

print("Kemasan ")
if in1[0]>0 :
    print("Kecil : "+str(in1[0]))
if in1[1]>0 :
    print("Besar  : "+ str(in1[1]))
```

    2. Proses

```python
x_proses = np.arange(-10, 1, 1)

proses_lb = fuzz.trapmf(x_proses, [-10, -10, -10, 1])
proses_cp = fuzz.trapmf(x_proses, [-10, 1, 1, 1])

fig, ax = plt.subplots(nrows=1, figsize=(7, 4))

ax.plot(x_proses, proses_lb, 'b', linewidth=1.5, label='lambat')
ax.plot(x_proses, proses_cp, 'g', linewidth=1.5, label='cepat')

ax.set_title('Proses')
ax.legend()

plt.tight_layout()
```

```python

proses = -6
in2 = []
in2.append(fuzz.interp_membership(x_proses, proses_lb, proses))
in2.append(fuzz.interp_membership(x_proses, proses_cp, proses))

print("Proses ")
if in2[0]>0 :
    print("Lambat : "+str(in2[0]))
if in2[1]>0 :
    print("Cepat  : "+ str(in2[1]))
```

    3. Harga

```python
x_harga = np.arange(30000, 300000, 1)

harga_rd = fuzz.trapmf(x_harga, [30000, 30000, 30000, 300000])
harga_tg = fuzz.trapmf(x_harga, [30000, 300000, 300000, 300000])

fig, ax = plt.subplots(nrows=1, figsize=(7, 4))

ax.plot(x_harga, harga_rd, 'b', linewidth=1.5, label='Rendah')
ax.plot(x_harga, harga_tg, 'g', linewidth=1.5, label='Tinggi')

ax.set_title('harga')
ax.legend()

plt.tight_layout()
```

Tinggi = 300000 Rendah = 30000

    [Tinggi] = 0; x ≤ 30000
             = 1; x = 300000
             = ( x – 30000) / (300000 – 30000 )
    
    [Rendah] = 0; x ≥ 300000
             = 1; x = 30000
             = (300000 – x ) / (300000 – 30000 )

    R3 IF Kemasan Kecil AND Proses Cepat THAN Harga Kopi Tinggi
